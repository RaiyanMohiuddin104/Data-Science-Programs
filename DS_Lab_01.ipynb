```python
#program to find linear search
def linearSearch(arr,target):
  for i in range(len(arr)):
    if arr[i] ==  target:
      return i;
  return -1
arr = [10,20,30,40]
target = int(input("Enter the element to search:- "))
result = linearSearch(arr,target)
if result != -1:
  print(f"Element {target} found at index {result}")
else:
  print(f"Element {target} not found")
```

```python
#program to find binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
numbers = [10, 20, 30, 40, 50]
target = int(input("Enter the number to be searched:-"))
result = binary_search(numbers, target)
if result != -1:
    print(f"Element {target} found at index {result}")
else:
    print(f"Element {target} not found")
```

```python
#program to find the mean,median and mode using import statistics method
import statistics
data = [101,102,103,104,104,105]
mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)
print(f"Mean of the data is {mean}")
print(f"Median of the data is {median}")
print(f"Mode of the data is {mode}")
```

```python
#Data Analysis using numpy
import numpy as np
data = np.array([14,25,34,46,55,62,77,86,94])
Sum = np.sum(data)
Average = np.mean(data)
Maximum = np.max(data)
Minimum = np.min(data)
print(f"Sum of the given data is:- {Sum}")
print(f"Average of the given data is:- {Average}")
print(f"Maximum of the given data is:- {Maximum}")
print(f"Minimum of the given data is:- {Minimum}")
```

```python
#program to perform bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted array:", bubble_sort(arr))
```
