```python
#13. Implementation of mathematical operations on pandas dataframe
import pandas as pd
data = {
    "Ages" : [22, 25, 47, 35, 46, 50, 28, 33, 39, 41],
    "Salaries": [3000, 3200, 5500, 4000, 5200, 6000, 3500, 3700, 4500, 4800],
}
df=pd.DataFrame(data)
print("Mean of the data is - ",df.mean())
print("Median of the data is - ",df.median())
print("Mode of the data is - ",df.mode())
print("Standard deviation of the data is - ",df.std())
print("Kurtosis of the data is - ",df.kurt())
print("Skewness of the data is - ",df.skew())
```

```python
import pandas as pd

url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
data = pd.read_csv(url)
print("Origial Data:- \n")
print(data.head())
print("\n")

#Data Filtering
filtered_data = data[data[' Height(Inches)"']>70]
print("Filtered Data:- \n")
print(filtered_data)
print("\n")

#Data Mnipulation
data['Height_cm'] = data[' Height(Inches)"'] * 2.54
print("Insertion of new column (Data Manipulation):- \n")
print(data.head())
print("\n")

#Data Ranking
data['Weight_rank'] = data[' "Weight(Pounds)"'].rank(ascending=False)
print("Ranked Data:- \n")
print(data.sort_values('Weight_rank').head())
print("\n")

```
